```python
# correlation matrix
df_corr = train.corr()
df_corr = df_corr.sort_values(by=['SalePrice'], axis=1)
df_corr = df_corr.sort_values(by=['SalePrice'], axis=0)
trace = go.Heatmap(x=df_corr.columns, y=df_corr.columns, z=df_corr.values, colorscale='Viridis', colorbar=dict(
                title='correlation coefficient'
            ))

data=[trace]

layout = go.Layout(
    title='Correlation Matrix',
    autosize=False,
    width=900,
    height=900,
    margin=go.layout.Margin(
        l=150,
        r=0,
        b=200,
        t=100,
        pad=4
    )
)

fig = dict(data=[trace], layout=layout)
py.iplot(fig)
```

```python
columns = ['SalePrice', 'OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath', 'YearBuilt']
dimensions = []
for column in columns:
    dimension = {
        "label": column,
        "values": train[column]
    }
    dimensions.append(dimension)

trace = go.Splom(dimensions=dimensions,
                
                marker=dict(
                            size=7,
                            showscale=False,
                            line=dict(width=0.5,
                                      color='rgb(230,230,230)')),
                 diagonal=dict(visible=False)
                )
```

```python

layout = go.Layout(
    title='Scatter Pairplot',
    dragmode='select',
    width=1000,
    height=1000,
    autosize=False,
    hovermode='closest',
    plot_bgcolor='rgba(240,240,240, 0.50)',
     
)

fig1 = dict(data=[trace], layout=layout)
py.iplot(fig1, filename='splom-iris1')
```
